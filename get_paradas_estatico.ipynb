```python
def busca_bairro(lat,lng):
    bairro = ''
    
    google_resp = requests.get('https://maps.googleapis.com/maps/api/geocode/json?latlng={},{}&key={}'.format(lat,lng,YOUR_API_KEY))
    #print(google_resp.content)

    if google_resp.status_code != 200:
        # This means something went wrong.
        raise ApiError('GET /tasks/ {}'.format(google_resp.status_code))
        return ''
    
    geo_info = google_resp.json()
    
    for component in geo_info['results'][0]['address_components']:
        if('sublocality_level_1' in component['types']):
            print(component)
            bairro = component['long_name']
    return bairro
```

```python
for parada in paradas:
    lat = parada['latitude']
    lng = parada['longitude']
    parada['bairro'] = busca_bairro(lat,lng)
```

```python
count_nao_processados = 0
idx_nao_processados = []
idx_processados = []
for idx,parada in enumerate(paradas):
    if('bairro' in parada.keys()):
        idx_processados.append(idx)
        continue;
    else:
        idx_nao_processados.append(idx)
        count_nao_processados += 1
        
        
print('count nao enriquecidos com bairro : {} de {}'.format(count_nao_processados,len(paradas)))
print('count paradas prontas para inserção no elastic : {}'.format(len(idx_processados)))
#print('idx nao processados : {}'.format(idx_nao_processados))
```

```python
for idx_parada in idx_nao_processados:
    parada = paradas[idx_parada]
    lat = parada['latitude']
    lng = parada['longitude']
    parada['bairro'] = busca_bairro(lat,lng)
```
